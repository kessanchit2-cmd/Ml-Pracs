```python
import numpy as np
```

```python
#input dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
```

```python
#orgate output
y = np.array([0,1,1,1])
```

```python
#initlizae weights and bais
hts = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10
```

```python
#training preceptron
for epoch in range(epochs):

    print("\nEpoch:", epoch+1)

    for i in range(len(X)):

        # Weighted Sum
        net_input = np.dot(X[i], weights) + bias

        # Activation Function
        prediction = 1 if net_input > 0 else 0

        # Error
        error = y[i] - prediction

        # Weight Update
        weights += learning_rate * error * X[i]

        # Bias Update
        bias += learning_rate * error

        print("Input:", X[i],
              "Target:", y[i],
              "Prediction:", prediction)
```

```python
#final output
print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```

```python
#Testing
print("\nTesting OR Gate")

for i in range(len(X)):

    net_input = np.dot(X[i], weights) + bias

    prediction = 1 if net_input > 0 else 0

    print(X[i], "->", prediction)
```
