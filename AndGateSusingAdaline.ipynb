```python
import numpy as np
```

```python
#input dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
```

```python
#And Gate Output
y = np.array([0,0,0,1])
```

```python
#initliaze parameter
weights = np.zeros(2)

bias = 0

learning_rate = 0.1

epochs = 20
```

```python
#Training adaline
for epoch in range(epochs):

    print("\nEpoch:", epoch+1)

    total_error = 0

    for i in range(len(X)):

        # Net Input
        net_input = np.dot(X[i], weights) + bias

        # Error
        error = y[i] - net_input

        # Weight Update
        weights = weights + learning_rate * error * X[i]

        # Bias Update
        bias = bias + learning_rate * error

        # Mean Squared Error
        total_error += error**2

        print("Input:", X[i],
              "Target:", y[i],
              "Net Input:", round(net_input,2),
              "Error:", round(error,2))

    print("MSE:", round(total_error/len(X),4))

```

```python
#testing
print("\nFinal Weights:", weights)

print("Final Bias:", bias)

print("\nTesting AND Gate")

for i in range(len(X)):

    net_input = np.dot(X[i], weights) + bias

    # Activation Function
    prediction = 1 if net_input >= 0.5 else 0

    print(X[i], "->", prediction)
```
