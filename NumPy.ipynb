```python
# ==========================================
# NumPy All Basic Operations
# ==========================================

import numpy as np

# ------------------------------------------
# Create an array of n elements
# ------------------------------------------

a = np.array([10,20,30,40,50,60,70,80,90])

print("Original Array:")
print(a)

# ------------------------------------------
# Print the 1st element
# ------------------------------------------

print("\n1st Element:")
print(a[0])

# ------------------------------------------
# Print the element at index 3
# ------------------------------------------

print("\nElement at Index 3:")
print(a[3])

# ------------------------------------------
# Print last element using positive indexing
# ------------------------------------------

print("\nLast Element using Positive Index:")
print(a[len(a)-1])

# ------------------------------------------
# Print last element using negative indexing
# ------------------------------------------

print("\nLast Element using Negative Index:")
print(a[-1])

# ------------------------------------------
# Output of a[4]
# ------------------------------------------

print("\nOutput of a[4]:")
print(a[4])

# ------------------------------------------
# Print first 3 elements
# ------------------------------------------

print("\nFirst 3 Elements:")
print(a[:3])

# ------------------------------------------
# Print elements from index 2 to 5
# ------------------------------------------

print("\nElements from Index 2 to 5:")
print(a[2:5])

# ------------------------------------------
# Print elements from index 3 to end
# ------------------------------------------

print("\nElements from Index 3 to End:")
print(a[3:])

# ------------------------------------------
# Print all elements except 1st one
# ------------------------------------------

print("\nAll Elements Except 1st:")
print(a[1:])

# ------------------------------------------
# Output of a[:4]
# ------------------------------------------

print("\nOutput of a[:4]:")
print(a[:4])

# ------------------------------------------
# Print every 2nd element
# ------------------------------------------

print("\nEvery 2nd Element:")
print(a[::2])

# ------------------------------------------
# Print every 3rd element
# ------------------------------------------

print("\nEvery 3rd Element:")
print(a[::3])

# ------------------------------------------
# Print elements 1 to 5 with step 2
# ------------------------------------------

print("\nElements 1 to 5 with Step 2:")
print(a[1:6:2])

# ------------------------------------------
# Reverse the array
# ------------------------------------------

print("\nReversed Array:")
print(a[::-1])

# ------------------------------------------
# Print every 2nd element in reverse order
# ------------------------------------------

print("\nEvery 2nd Element in Reverse:")
print(a[::-2])

# ------------------------------------------
# Elements greater than 30
# ------------------------------------------

print("\nElements Greater Than 30:")
print(a[a > 30])

# ------------------------------------------
# Even numbers
# ------------------------------------------

print("\nEven Numbers:")
print(a[a % 2 == 0])

# ------------------------------------------
# Odd numbers
# ------------------------------------------

print("\nOdd Numbers:")
print(a[a % 2 != 0])

# ==========================================
# Mathematical Operations
# ==========================================

print("\nSum:")
print(np.sum(a))

print("\nMean:")
print(np.mean(a))

print("\nMaximum:")
print(np.max(a))

print("\nMinimum:")
print(np.min(a))

print("\nStandard Deviation:")
print(np.std(a))

print("\nSquare Root:")
print(np.sqrt(a))

# ==========================================
# Matrix Operations
# ==========================================

m1 = np.array([[1,2],[3,4]])

m2 = np.array([[5,6],[7,8]])

print("\nMatrix 1:")
print(m1)

print("\nMatrix 2:")
print(m2)

# Addition
print("\nMatrix Addition:")
print(m1 + m2)

# Subtraction
print("\nMatrix Subtraction:")
print(m1 - m2)

# Multiplication
print("\nMatrix Multiplication:")
print(np.dot(m1, m2))

# Transpose
print("\nTranspose of Matrix 1:")
print(m1.T)
```
