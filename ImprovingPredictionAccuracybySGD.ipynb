```python
import numpy as np
```

```python
#training Data
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
```

```python
#OrGates
y = np.array([0,1,1,1])
```

```python
#initizale the weights
weights = np.zeros(2)

bias = 0

learning_rate = 0.1

epochs = 10
```

```python
#Activation function
def activation(x):

    if x >= 0:
        return 1
    else:
        return 0
```

```python
#training using SGD
for epoch in range(epochs):

    print("\nEpoch:", epoch+1)

    for i in range(len(X)):

        # Weighted Sum
        net_input = np.dot(X[i], weights) + bias

        # Prediction
        prediction = activation(net_input)

        # Error
        error = y[i] - prediction

        # SGD Weight Update
        weights = weights + learning_rate * error * X[i]

        # Bias Update
        bias = bias + learning_rate * error

        print("Input:", X[i],
              "Target:", y[i],
              "Prediction:", prediction,
              "Error:", error)
```

```python
#Final Weights
print("\nFinal Weights:", weights)

print("Final Bias:", bias)
```

```python
#testing
print("\nTesting Perceptron")

for i in range(len(X)):

    net_input = np.dot(X[i], weights) + bias

    prediction = activation(net_input)

    print(X[i], "->", prediction)
```
